```python
question = ['Q2', 'Q3', 'Q4', 'Q5', 'Q6_1', 'Q6_2', 'Q6_3', 'Q6_4', 'Q6_5', 'Q6_6',
            'Q6_7', 'Q6_8', 'Q6_9', 'Q6_10', 'Q6_11', 'Q6_12', 'Q7_1', 'Q7_2',
            'Q7_3', 'Q7_4', 'Q7_5', 'Q7_6', 'Q7_7', 'Q8', 'Q9', 'Q10_1', 'Q10_2',
            'Q10_3']

single_choice = [] #단일 문항만을 추려낼 것이기 때문에 우선 비어있는 리스트를 생성함

for q in question: #question 리스트의 범위를 순회하도록 시행함
    if '_' not in q: #단일 문항을 추려내기 위한 조건
        single_choice.append(q) #단일 문항이면 맨 처음에 생성한 리스트에 추가하도록 함
print(single_choice)
```

## Q2.

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/healthexp.csv')

df_filtered = df[df['Year'] >= 2011] #2011년을 포함한 연도들을 필터링함
Life_Exp = df_filtered.groupby(['Year', 'Country'])['Life_Expectancy'].mean().unstack() #연도별, 국가별로 그룹화 하여 평균 기대수명을 구함

Life_Exp #데이터프레임 형태로 출력
```

### Python 속성 코스

- **튜플은 iterable한 개체이지만 immutable하다! (list와의 차이점)**
- **for문과 while문의 차이 익히기**: for문은 designated 범위 내를 순회, while문은 조건이 참일 때까지 무한 반복 시행

### Pandas로 파일 불러오기 및 기초 익히기
- **CSV 파일 저장**: `to_csv("파일명", index=False)`
- **CSV 파일 불러오기**: `read_csv("파일명")`
- **행과 열의 수 보기**: `shape`
- **일부 데이터 가져오기**: `head()`, `tail()`, `sample()`
- **요약과 기술통계 값 구하기**: `info()`, `describe()`, `nunique()`, `index`, `columns`, `values`
- **CSV 파일 저장**: df.to_csv('output.csv', index=False)

### Pandas의 DataFrame과 Series 이해하기
- **Series**: 1차원 벡터구조 (not framed)
- **DataFrame**: 2차원 행렬구조

### 색인하기
- **컬럼 색인**: `df['컬럼명']`
- **행 색인**: `df.loc[행 인덱스]`
- **행과 열 색인**: `df.loc[행 인덱스, 열 이름]`
- **조건식 색인**: `df.loc[조건식, 열 이름]`
- **빠른 접근을 위해 색인함!**

### DataFrame 다루기
- **열 인덱싱**: `df['컬럼명']`
- **행 인덱싱**: `df.loc[행 인덱스]`
- **행, 열 인덱싱**: `df.loc[행 인덱스, 열 이름]`
- **정렬하기**: `df.sort_values(by='컬럼명')`
- **조건식 사용하기**: `df[df['컬럼명'] > 값]`
- **빈도수 구하기**: `df['컬럼명'].value_counts()`

### groupby와 pivot_table
- **groupby**: 데이터 그룹화 후 다양한 연산 수행
- **pivot_table**: 데이터 피벗 테이블 생성
- **그룹화**: grouped_df = df.groupby('column_name').mean()
- **피벗 테이블 생성**: pivot_df = df.pivot_table(values='value', index='index_col', columns='columns_col', aggfunc='mean')
